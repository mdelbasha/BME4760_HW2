Dataset description: This dataset is composed of a range of biomedical voice measurements from patients with Parkinson's disease (PD). Each column in the table is a particular voice measure, and each row corresponds one of 195 voice recording from these individuals ("name" column). The main aim of the data is to discriminate healthy people from those with PD, according to "status" column which is set to 0 for healthy and 1 for PD.*

```python
import numpy as np
import pandas as pd
```

```python
# Get csv file and upload it
raw_data = pd.read_csv('data.csv')
raw_data
```

# Homework Part b
**Goal:** *Split the dataset for training a knn classifier in a holdout + cross-validation fashion. Run the k-nn (k-nearest neighbors) algorithm on this dataset and find the optimal k using Grid search, where k={4, 5, 6, 7, 8, 9, 10} (Hint: use the GridSearchCV operator in sklearn).[10 points]*

**Note:** Below we are loading in specific classes from the *SciKit-learn* library, which is an open source software library for machine learning on python.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import train_test_split
```

## Using Holdout to create Train and Test Data
Below the uploaded data is being split into training and testing data. First the voice recordings are being extracted from the data and then the status column is being selected to obtain the label for whether or not the inividuals are healthy or exhibit Parkingson's Disease.

```python
# set up values and label
X_voice = raw_data.iloc[:, 1:23].values  # All patient voice recording values
Y_status = raw_data.iloc[:,-1].values  # last column of the data frame raw_data
```

```python
# Make training and test data
X_train, X_test, y_train, y_test = \
    train_test_split(X_voice,
                     Y_status,
                     test_size=0.20,
                     stratify=Y_status,
                     random_state=1)
```

```python
# Explore the dimensions of training and testing data.
print("X_train shape: ", X_train.shape)
print("Y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("Y_test shape: ", y_test.shape)
```

## Create Pipeline
A pipeline is a container that allows one to perform a sequence of different transforms on data. Here we will create a pipeline to allow us to use the K-Nearest Neighbors algorithm on our data set of voice recordings.

```python
# Make a pipeline
pipe = make_pipeline(StandardScaler(),
                     PCA(n_components=2),
                     KNeighborsClassifier())
```

## Perform Grid Search
Grid search is a method in machine learning which allows for the combination of hyperparameter tuning and cross validations in one step. We use the **GridSearchCV** class in **sklearn.model_selection** package from scikit-learn.

```python
# Performing Grid Search
param_range = list(range(4,11))
param_grid = [{'kneighborsclassifier__n_neighbors' : param_range}]

grid = GridSearchCV(estimator=pipe, 
                  param_grid=param_grid, 
                  scoring='accuracy', 
                  cv=10,
                  n_jobs=-1)

grid_eval = grid.fit(X_train, y_train)
print(grid_eval.best_score_)
print(grid_eval.best_params_)
```

```python
# Test the above trained model from gridsearch optimization
acc_measure = grid_eval.best_estimator_
acc_measure.fit(X_train, y_train)
print('Test accuracy with gridsearch: %.2f' % acc_measure.score(X_test,y_test))
```

**NOTE** Given the random seed set for holdout, and using 10-fold cross validation, the best K values for our knn was 5 which produced a test accuracy of 85%.

# Homework Part c
**Goal:** *Plot the learning curve, as well as the validation curve [6 points].*

## Creating Learning Curve
A learning curve is a plot depicting improvement in performance plotted against sample size.

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores =\
                learning_curve(estimator=pipe,
                               X=X_train,
                               y=y_train,
                               train_sizes=np.linspace(0.1, 1.0, 10),
                               cv=10,
                               n_jobs=1)

# Obtain mean and standard deviation on train and test scores
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
```

```python
# Put the training accuracy on the plot
plt.plot(train_sizes,
         train_mean,
         color='blue',
         marker='o',
         markersize=5,
         label='training accuracy')

plt.fill_between(train_sizes,
                 train_mean + train_std,
                 train_mean - train_std,
                 alpha=0.15,
                 color='blue')

# Put the validation accuracy on the plot
plt.plot(train_sizes,
         test_mean,
         color='green',
         linestyle='-.',
         marker='s',
         markersize=5,
         label='Validation Accuracy')

plt.fill_between(train_sizes,
                 test_mean + test_std,
                 test_mean - test_std,
                 alpha=0.15,
                 color='green')

plt.grid()
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

## Validation Curve
The validation curve also plots the performance of a model but against changes in parameters.

```python
from sklearn.model_selection import validation_curve

train_scores, test_scores = validation_curve(
                estimator=pipe, 
                X=X_train, 
                y=y_train, 
                param_name='kneighborsclassifier__n_neighbors', 
                param_range=param_range,
                cv=10)

# Obtain mean and standard deviation on train and test scores
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
```

```python
# Put the training accuracy on the plot
plt.plot(param_range,
         train_mean, 
         color='blue',
         marker='o', 
         markersize=5,
         label='training accuracy')

plt.fill_between(param_range,
                 train_mean + train_std,
                 train_mean - train_std,
                 alpha=0.15,
                 color='blue')

# Put the validation accuracy on the plot
plt.plot(param_range,
         test_mean, 
         color='green',
         linestyle='-.', 
         marker='s',
         markersize=5, 
         label='validation accuracy')

plt.fill_between(param_range, 
                 test_mean + test_std,
                 test_mean - test_std, 
                 alpha=0.15,
                 color='green')

plt.grid()
plt.legend(loc='lower right')
plt.xlabel('N nearest neighbors')
plt.ylabel('Accuracy')
plt.ylim([0.6, 1.0])
plt.tight_layout()
plt.show()
```
